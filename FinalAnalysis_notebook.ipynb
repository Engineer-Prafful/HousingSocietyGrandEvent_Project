### Replacing '-' with NULL value

```python
df.replace('-',np.nan,inplace = True)
df
```

```python
df.isna().sum()
```

### Droping the unwanted columns

```python
df.drop(['Unnamed: 0','Owner\'s Spouse Name'], axis = 1, inplace = True)
df.head()
```

### Changing the datatype to numeric

```python
df['No of Resident'] = pd.to_numeric(df['No of Resident'], errors='coerce')
df['Confirmed Members'] = pd.to_numeric(df['Confirmed Members'], errors='coerce')
df['Maintenance Amt'] = pd.to_numeric(df['Maintenance Amt'], errors = 'coerce')
```

### Impute missing values in 'Maintenance Amt' based on flat_area

```python
avg_maintenance_amt = {500:5500,550:6000,600:7000}
```

```python
for index, row in df.iterrows():

    if pd.isnull(row['Maintenance Amt']):

        flat_area = row['Flat Area (sq.mt)']

        if flat_area in avg_maintenance_amt:

            df.at[index, 'Maintenance Amt'] = avg_maintenance_amt[flat_area]
```

```python
df.head()
```

### Impute missing values in 'Flat Vacancy' column according to Owner Availability

```python
df['Flat Vacancy'].fillna(df['Availlability of owner'].apply(lambda x: 'Owned' if x == 'Yes' else 'Vacant'), inplace=True)
```

```python
df.head()
```

### Impute missing values in 'Donation', 'no of resident', 'confirmed_member' column with the median

```python
df['Donation'].fillna(df['Donation'].median(),inplace = True)
df['No of Resident'].fillna(df['No of Resident'].median(),inplace = True)
df['Confirmed Members'].fillna(df['Confirmed Members'].median(),inplace = True)
```

```python
df.head()
```

###  Impute missing values in 'Availability of owner', 'origin of owner'  with mode

```python
df['Availlability of owner'].fillna(df['Availlability of owner'].mode()[0], inplace = True)
df['Origin of Owner'].fillna(df['Origin of Owner'].mode()[0], inplace = True)
```

```python
df.head()
```

```python
 df.isna().sum()
```

## Calculating Outsiders

```python
df['Outsiders'] = df['Confirmed Members'] - df['No of Resident']
for index,row in df.iterrows():
    if row['Outsiders'] <= 0:
        df.at[index,'Outsiders'] = 0
        
df.head()
```

### Applying the appropriate cases

```python
df['Origin of Owner'] = df['Origin of Owner'].apply(lambda x : x.lower())
df['Tenant or owner'] = df['Tenant or owner'].apply(lambda x : x.lower())
df['Availlability of owner'] = df['Availlability of owner'].apply(lambda x : x.lower())
df['Flat Vacancy'] = df['Flat Vacancy'].apply(lambda x : x.lower())

df['Owner Name'] = df['Owner Name'].apply(lambda x : x.title())

df.head()
```
